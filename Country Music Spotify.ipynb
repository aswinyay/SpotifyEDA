```python
import requests
import json
import pandas as pd

CLIENT_ID = '9444e5d0012f4b2688a29c093833655c'
CLIENT_SECRET = '87fe2061e440456fa5dafed6388b672d'
```

```python
AUTH_URL = 'https://accounts.spotify.com/api/token'

# POST
auth_response = requests.post(AUTH_URL, {
    'grant_type': 'client_credentials',
    'client_id': CLIENT_ID,
    'client_secret': CLIENT_SECRET,
})

# convert the response to JSON
auth_response_data = auth_response.json()

# save the access token
access_token = auth_response_data['access_token']
```

```python
headers = {
    'Authorization': 'Bearer {token}'.format(token=access_token)
}
```

```python
playlist_id = '1lcbYKyLFvyLb5rZXbVVYc?si=704cc3c316604b57'
```

```python
BASE_URL = 'https://api.spotify.com/v1/playlists/'
```

```python
track_dict = []
```

```python
r = requests.get(BASE_URL + playlist_id + '/tracks', headers=headers)
track_dict = r.json()['tracks']['items']
```

```python
track_list = []
for t in track_dict:
    track_list.append(t['track']['id'])
```

```python
TRACK_BASE_URL = 'https://api.spotify.com/v1/audio-features/'
track_audio_features = []
for audio in track_list: 
    f = requests.get(TRACK_BASE_URL  + audio,headers=headers)
    f = f.json()
    track_audio_features.append(f)
```

```python
track_af_df = pd.DataFrame(track_audio_features)
track_af_df
```
